# 4. Feature Scaling

````python
from sklearn.preprocessing import StandardScaler
````

````python
df2 = df.copy()
df2.drop('Private',axis=1, inplace=True)
````

````python
df_list = list(df2)
df_list
````

````python
scaler = StandardScaler()
df2[df_list] = scaler.fit_transform(df2[df_list])
````

````python
df2.head()
````

# 5. Determine the Number of Cluster

````python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
````

## Elbow Analysis

````python
WCSS = []
for i in range (2,12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df2)
    WCSS.append(kmeans.inertia_)

plt.plot(range(2,12), WCSS)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
````

## Silhouette Score

````python
score = []
for k in range (2,25):
    kmeans = KMeans(n_clusters=k)
    cluster = kmeans.fit_predict(df2)
    siluet = silhouette_score(df2, cluster, metric='euclidean')
    score.append(siluet)

plt.figure(figsize=(10,5))
plt.plot(range(2,25), score, 'r-')
plt.ylabel ('Silhouette')
plt.xlabel ('n_cluster');
````

```json
Based on the Silhouette Score and Elbow Analysis, I decided to use 3 clusters.
```

# 6. Modeling

````python
model = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
clusters = model.fit_predict(df2)
````

````python
model.cluster_centers_
````

````python
df_pred = df.copy()
df_pred['Cluster'] = clusters
df_pred.head()
````

````python
tsne = TSNE(2, random_state = 42)
X_train_tsne = tsne.fit_transform(X_train)
````
